## Handling Missing Values

```python
!wget https://raw.githubusercontent.com/krishnaik06/Pyspark-With-Python/main/test2.csv
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Practise').getOrCreate()
```

```python
df_pyspark = spark.read.csv('test2.csv', header=True, inferSchema=True)
```

```python
df_pyspark.show()
```

```python
# Drop Columns
df_pyspark.drop('Name').show()
```

```python
# Delete rows with missing values (null)
df_pyspark.na.drop().show()
```

```python
# When how='all', it deletes rows will all null values
df_pyspark.na.drop(how='all').show()
```

```python
# Thresholding - deletes rows when the amount of null values of a row is greater the thresh value
df_pyspark.na.drop(how='all', thresh = 2).show()
```

```python
# Filling the Missing Values
df_pyspark.na.fill('Missing Values', 'Name').show()
```

```python
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=['age', 'Experience', 'Salary'], 
    outputCols=["{}_imputed".format(c) for c in ['age', 'Experience', 'Salary']]
    ).setStrategy("median")
```

```python
# Add imputation cols to df
imputer.fit(df_pyspark).transform(df_pyspark).show()
```

## Filter Operations

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('dataframe').getOrCreate()
```

```python
df_pyspark = spark.read.csv('test1.csv', header=True, inferSchema=True)
df_pyspark.show()
```

```python
# Salary of people less than or equal to 20,000 salary
df_pyspark.filter('Salary<=20000').show()
```

```python
# Salary of ONLY the Name and age of people less than or equal to 20,000 salary
df_pyspark.filter('Salary<=20000').select(['Name', 'age']).show()
```

```python
# Show salaries between 15000 and 20000
df_pyspark.filter((df_pyspark['Salary']<=20000) &
                  (df_pyspark['Salary']>=15000)).show()
```

```python
# NOT FUNCTION - showing salaries that are not less than 20000
df_pyspark.filter(~(df_pyspark['Salary']<=20000)).show()
```

## GroupBy & Aggregate Functions

```python
!wget https://raw.githubusercontent.com/krishnaik06/Pyspark-With-Python/main/test3.csv
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Agg').getOrCreate()
```

```python
spark
```

```python
df_pyspark = spark.read.csv('test3.csv', header=True, inferSchema=True)
```

```python
df_pyspark.show()
```

```python
df_pyspark.printSchema()
```

```python
# Groupby - shows the total salary of each person
df_pyspark.groupBy('Name').sum().show()
```

```python
# Shows the people with the maximum salary
df_pyspark.groupBy('Name').max().show()
```

```python
# Shows the people with the min salary
df_pyspark.groupBy('Name').min().show()
```

```python
# Group by departments - total salary of each department
df_pyspark.groupBy('Departments').sum().show()
```

```python
# Group by departments - average salary of each department
df_pyspark.groupBy('Departments').mean().show()
```

```python
# Group by departments - the number of people in each department
df_pyspark.groupBy('Departments').count().show()
```

```python
# Total salary being given
df_pyspark.agg({'Salary':'sum'}).show()
```

## Using PySpark MLib

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Missing').getOrCreate()
```

```python
# Read the dataset
training = spark.read.csv('test1.csv', header=True, inferSchema=True)
```

```python
training.show()
```

```python
training.printSchema()
```

```python
training.columns
```

```python
# Group the features needed to give a predictions
from pyspark.ml.feature import VectorAssembler
featureassembler = VectorAssembler(inputCols=['age', 'Experience'],
                                   outputCol='Independent Features')
```

```python
output = featureassembler.transform(training)
```

```python
output.show()
```

```python
output.columns
```

```python
# Make new dataframe for the inputs and outputs
finalised_data = output.select('Independent Features', 'Salary')
```

```python
finalised_data.show()
```

```python
from pyspark.ml.regression import LinearRegression
# Train and test split
train_data, test_data = finalised_data.randomSplit([0.75, 0.25])
regressor = LinearRegression(featuresCol='Independent Features', labelCol='Salary')
regressor = regressor.fit(train_data)
```

```python
# coefficients
regressor.coefficients
```

```python
# intercepts
regressor.intercept
```

```python
# Prediction
pred_results = regressor.evaluate(test_data)
```

```python
pred_results.predictions.show()
```

```python
# Mean Absolute Error and Mean Squared Error
pred_results.meanAbsoluteError, pred_results.meanSquaredError
```
